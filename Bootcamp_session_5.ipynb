```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
print(df["year"].min(),df["year"].max())
print(len(df["country"].unique()))
print(len(df["continent"].unique()))
```

```python
df= df.rename(columns = {
    "pop":"population"
})
```

```python
print(df["continent"].value_counts(()))
```

```python
#.mean().sort_values(ascending="False")
```

```python
#
life_bar=df.groupby("continent")["lifeExp"].mean().sort_values()
plt.figure(figsize=(8,5))
plt.bar(life_bar.index,life_bar.values)
plt.title("life expectancy by cotinent")
plt.xlabel("continent")
plt.ylabel("life_exp")
plt.xtick(rotation=45)
plt.tight_layout()
plt.show()
```
